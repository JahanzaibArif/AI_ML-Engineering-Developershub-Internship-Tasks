# Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the Iris Dataset

```python
# Load the Iris dataset
iris = sns.load_dataset("iris")
```

# Inspect the Dataset

```python
# Shape of the dataset
print("Shape:", iris.shape)
```

```python
# Column names
print("\nColumns:", iris.columns.tolist())
```

```python
# First few rows
print("\nHead:\n", iris.head())
```

# Summary Information

```python
# Info about data types and missing values
print("\nInfo:")
print(iris.info())
```

```python
# Summary statistics
print("\nDescribe:")
print(iris.describe())
```

# Visualizations

### Scatter Plot

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

### Histograms

```python
iris.hist(edgecolor='black', figsize=(10, 8))
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

### Box Plots

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=iris, x="species", y=column)
    plt.title(f"{column.capitalize()} by Species")
plt.tight_layout()
plt.show()
```
